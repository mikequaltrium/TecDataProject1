```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Resources/New datasets/urban-vs-rural-majority.csv")
df.head()
```

```python
# df.drop(['Code','Rural (%)'],axis=1)
countries = df["Entity"].unique()
years = np.arange(1950,2019,1)
```

```python
new_df = pd.DataFrame({"Countries" : countries})
new_df = new_df.set_index('Countries')

countries
for cnt in countries:
    for yr in years:
        try:
            val = df.loc[(df["Entity"] == cnt) & (df["Year"] == yr), "Urban (%)"].values[0]
            if(val == 0):
                val = np.NaN
            new_df.at[cnt, yr] = "{0:.2f}".format(val)
        except:
            new_df.at[cnt, yr] = np.NaN

new_df = new_df.replace('nan', np.NaN)
new_df = new_df.dropna(how='all')
new_df
```

```python
new_df.to_csv('Resources/urban-vs-rural-majority-clean.csv')
```
