```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Resources/global_mortality.csv")
print(df.columns)

# df = df.set_index('country')
```

```python
filter_df = df[['country','year','Dementia (%)','Cardiovascular diseases (%)','Diabetes (%)','Nutritional deficiencies (%)','Protein-energy malnutrition (%)']]
```

```python
# what are the countries with the highest mortality % by Dementia
sorted_by_dimentia_df = filter_df.sort_values(by='Dementia (%)', ascending=False)
sorted_by_dimentia_df = sorted_by_dimentia_df.set_index('country')
sorted_by_dimentia_df.head()
```

```python
years = sorted_by_dimentia_df['year'].unique()
countries = sorted_by_dimentia_df.index.unique()
top_countries = countries[0 :5].tolist()
top_countries.append("Mexico")
top_countries.append("United States")
top_countries.append("England")
```

```python
plt.title("% Mortality by Dimensia / Alzheimer By Country")
plt.ylabel("% Mortality")
plt.xlabel("Year")
plt.grid()

plt.ylim(0, 20)
plt.xlim(1989, 2019)

plt.xticks( np.arange(1989,2017, 1) )
plt.yticks(np.arange(0,20,1))

for country in top_countries:
    demensia_vals = sorted_by_dimentia_df.loc[country, ['Dementia (%)']]['Dementia (%)'].tolist()
    plt.plot(years, demensia_vals, label=country, alpha=0.8, linewidth=3)

plt.legend(loc="upper left",labels=top_countries)
plt.rcParams["figure.figsize"] = [18,8]
plt.show()
```

```python
plt.title("% Mortality by Cardiovascular diseases By Country")
plt.ylabel("% Mortality")
plt.xlabel("Year")
plt.grid()

plt.ylim(14,50)
plt.xlim(1989, 2019)

plt.xticks( np.arange(1989,2017, 1) )
plt.yticks(np.arange(14,50,2))

for country in top_countries:
    cardio_vals = sorted_by_dimentia_df.loc[country, ['Cardiovascular diseases (%)']]['Cardiovascular diseases (%)'].tolist()
    plt.plot(years, cardio_vals, label=country, alpha=0.8, linewidth=3)

plt.legend(loc="upper left",labels=top_countries)
plt.rcParams["figure.figsize"] = [18,8]
plt.show()
```

```python
plt.title("% Mortality by Nutritional deficiencies diseases By Country")
plt.ylabel("% Mortality")
plt.xlabel("Year")
plt.grid()

plt.ylim(0,5)
plt.yticks(np.arange(0,5,1))

plt.xlim(1989, 2019)
plt.xticks( np.arange(1989,2017, 1) )


for country in top_countries:
    nutritional_vals = sorted_by_dimentia_df.loc[country, ['Nutritional deficiencies (%)']]['Nutritional deficiencies (%)'].tolist()
    plt.plot(years, nutritional_vals, label=country, alpha=0.8, linewidth=3)

plt.legend(loc="upper left",labels=top_countries)
plt.rcParams["figure.figsize"] = [18,8]
plt.show()
```

```python
pais = "Mexico"
plt.title("% Mortality by Dimensia / Alzheimer In "+pais)
plt.ylabel("% Mortality")
plt.xlabel("Year")
plt.grid()

demensia_vals = sorted_by_dimentia_df.loc[pais, ['Dementia (%)']]['Dementia (%)'].tolist()
cardio_vals = sorted_by_dimentia_df.loc[pais, ['Cardiovascular diseases (%)']]['Cardiovascular diseases (%)'].tolist()
nutri_vals = sorted_by_dimentia_df.loc[pais, ['Nutritional deficiencies (%)']]['Nutritional deficiencies (%)'].tolist()


ax = plt.subplot(111)

xpos1 = years-0.2
xpos2 = years
xpos3 = years+0.2

ax.bar(xpos1, demensia_vals, width=0.2, color='b', align='center',label="'Dementia (%)'")
ax.bar(xpos2, cardio_vals, width=0.2, color='g', align='center',label="Cardiovascular diseases (%)")
ax.bar(xpos3, nutri_vals, width=0.2, color='r', align='center',label="Nutritional deficiencies (%)")

ax.legend(loc="best")
# plt.show()
# plt.ylim(0, 20)
# plt.xlim(1989, 2019)

# plt.xticks( np.arange(1989,2017, 1) )
# plt.yticks(np.arange(0,20,1))


# plt.ylim(0,5)
# plt.yticks(np.arange(0,5,1))

# plt.xlim(1989, 2019)
# plt.xticks( np.arange(1989,2017, 1) )


# ax.legend(loc="best")
# plt.rcParams["figure.figsize"] = [18,8]
```
