```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Resources/New datasets/work-hours-per-week.csv")
df.rename(columns={" (hours per week)":"hours"},inplace=True)
df
```

```python
countries = df["Entity"].unique()
years = df["Year"].unique()
```

```python
new_df = pd.DataFrame({"Countries" : countries})
new_df = new_df.set_index('Countries')

for cnt in countries:
    for yr in years:
        try:
            val = df.loc[(df["Entity"] == cnt) & (df["Year"] == yr), "hours"].values[0]
            prev_yr_val = val
            if val == 0 | val == 'NaN':
                val = prev_yr_val
            new_df.at[cnt, yr] = "{0:.2f}".format(val)
        except:
            new_df.at[cnt, yr] = prev_yr_val

new_df
```

```python
new_df.to_csv('Resources/work-hours-per-week-clean.csv')
```
